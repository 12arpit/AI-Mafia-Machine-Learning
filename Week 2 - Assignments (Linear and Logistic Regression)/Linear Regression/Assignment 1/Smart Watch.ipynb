```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfx=pd.read_csv('Training Data/Linear_X_Train.csv')
dfy=pd.read_csv('Training Data/Linear_Y_Train.csv')
dftx=pd.read_csv('Testing Data/Linear_X_Test.csv')
x=dfx.values
y=dfy.values
tx=dftx.values
x=x.reshape((-1,))
y=y.reshape((-1,))
tx=tx.reshape((-1,))
```

```python
plt.scatter(x, y)
plt.show()
```

```python
def hypothesis(x, theta):
    return theta[0]+theta[1]*x

def error(x, y, theta):
    m=x.shape[0]
    e=0
    for i in range(m):
        hx=hypothesis(x[i], theta)
        e+=(hx-y[i])**2
    return e

def gradient(x, y, theta):
    g=np.zeros((2,))
    m=x.shape[0]
    for i in range(m):
        hx=hypothesis(x[i], theta)
        g[0]+=hx-y[i]
        g[1]+=(hx-y[i])*x[i]
    return g

def gradient_dec(x, y, lr=0.0001):
    theta=np.array([-2.0, 0.0])
    itr=0
    maxitr=100
    error_list=[]
    theta_list=[]
    while(itr<=maxitr):
        g=gradient(x, y, theta)
        e=error(x, y, theta)
        error_list.append(e)
        
        theta_list.append((theta[0], theta[1]))
        theta[0]-=lr*g[0]
        theta[1]-=lr*g[1]
        itr+=1
    return theta, error_list, theta_list
```

```python
theta, error_l, theta_l=gradient_dec(x, y)
```

```python
plt.plot(error_l)
plt.show()
```

```python
plt.scatter(x, y ,label='data')
plt.plot(tx, hypothesis(tx, theta), color='orange', label='prediction')
plt.legend()
plt.show()
```
